# Data Generation as we won't use actuall dataset for elaboration

```python
import numpy as np
import pandas as pd

np.random.seed(0)

df = pd.DataFrame({
    "small_feature": np.random.rand(100),        # 0–1
    "medium_feature": np.random.rand(100) * 100, # 0–100
    "large_feature": np.random.rand(100) * 10000 # 0–10000
})

print(df.head())
X = df.values.astype(float)

```

We will make the following three methods for scaling [Min-Max Scaling , Mean Normalization (strandardization) , Robust Scaling]


# 1. Manual Implementation

**Min-Max**

```python
X_min = X.min(axis=0)
X_max = X.max(axis=0)
Min_Max_Scaled_X = (X - X_min) / (X_max - X_min)
print(Min_Max_Scaled_X[:5])
```

**Mean Normalization**

```python
X_Mean = X.mean(axis=0)
X_Std = X.std(axis=0)
Standard_Scaled_X = (X - X_Mean) / X_Std
print(Standard_Scaled_X[:5])
```

**Robust Scaling**

```python
X_Q1 = np.percentile(X, 25, axis=0)
X_Q3 = np.percentile(X, 75, axis=0)
X_Median = np.median(X, axis=0)
IQR = X_Q3 - X_Q1
Robust_Scaled_X = (X - X_Median) / IQR
print(Robust_Scaled_X[:5])
```

# 2.Using scikit-learn

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

**Min-Max**

```python
x_min_max_scaled = MinMaxScaler().fit_transform(X)
print(x_min_max_scaled[:5])
```

**Mean Normalization**

```python
x_standard_scaled = StandardScaler().fit_transform(X)
print(x_standard_scaled[:5])
```

**Robust Scaling**

```python
x_robust_scaled = RobustScaler().fit_transform(X)
print(x_robust_scaled[:5])
```

# 3.Pytorch 

```python
import torch
X_torch = torch.tensor(X, dtype=torch.float32)
```

**Min-Max**

```python
min_values, _ = torch.min(X_torch, dim=0)
max_values, _ = torch.max(X_torch, dim=0)
min_max_scaled_torch = (X_torch - min_values) / (max_values - min_values)
print(min_max_scaled_torch[:5])
```

**Mean Normalization**

```python
mean_values = torch.mean(X_torch, dim=0)
std_values = torch.std(X_torch, dim=0)
standard_scaled_torch = (X_torch - mean_values) / std_values
print(standard_scaled_torch[:5])
```

**Robust Scaling**

```python
q1_values = torch.quantile(X_torch, 0.25, dim=0)
q3_values = torch.quantile(X_torch, 0.75, dim=0)
median_values = torch.median(X_torch, dim=0).values
iqr_values = q3_values - q1_values
robust_scaled_torch = (X_torch - median_values) / iqr_values
```

# 4. TensorFlow

```python
import tensorflow as tf
X_tf = tf.constant(X, dtype=tf.float32)
```

**Min-Max**


```python
min_values_tf = tf.reduce_min(X_tf, axis=0)
max_values_tf = tf.reduce_max(X_tf, axis=0)
min_max_scaled_tf = (X_tf - min_values_tf) / (max_values_tf - min_values_tf)
print(min_max_scaled_tf[:5])
```

**Mean Normalization**


```python
mean_values_tf = tf.reduce_mean(X_tf, axis=0)
std_values_tf = tf.math.reduce_std(X_tf, axis=0)
standard_scaled_tf = (X_tf - mean_values_tf) / std_values_tf
print(standard_scaled_tf[:5])
```

**Robust Scaling**


```python
import tensorflow_probability as tfp

q1_values_tf = tfp.stats.percentile(X_tf, 25.0, axis=0)
q3_values_tf = tfp.stats.percentile(X_tf, 75.0, axis=0)
median_values_tf = tfp.stats.percentile(X_tf, 50.0, axis=0)
iqr_values_tf = q3_values_tf - q1_values_tf
robust_scaled_tf = (X_tf - median_values_tf) / iqr_values_tf
print(robust_scaled_tf[:5])
```
