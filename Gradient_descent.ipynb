# Gradient Descent

- Gradient Descent is an optimization algorithm used to find the local minimum of a function. It is used in machine learning to minimize a cost or loss function by iteratively updating parameters in the opposite direction of the gradient. It works by calculating the derivative i.e slope of a function and moving in the direction opposite to the slope to reach a minimum. It can be applied to any differentiable function.
- y = mx + b, while m is the parameter we want to find so we do iterations on it till convergance when the loss tends to zero
### Gradient Descent Formula

X_new = X_old - first derivative of loss func

**Intuition:**  
Gradient Descent updates \(x\) in the *opposite direction* of the gradient to minimize the function — moving step-by-step toward the local minimum.
# Example
Implementation
Here’s an example to find the minimum of 
f(x) =x^2 +4x+ 4
# **Solution**

```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def f(x):
    return x**2 + 4*x + 4
```

```python
def df(x):
    return 2*x + 4
```

# Gradient Descent function


```python
# Gradient Descent function
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    history = [x]  # store x values for plotting
    for i in range(iterations):
        x = x - learning_rate * df(x)
        history.append(x)
        print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
    return x, history 
```

```python
starting_point = 0
learning_rate = 0.1
iterations = 10

minimum, history= gradient_descent(starting_point, learning_rate, iterations)
print(f"\nLocal minimum occurs at x = {minimum:.4f}, f(x) = {f(minimum):.4f} , Num of iterations = {iterations}")
```

```python
# Visualization function
def visualize(history, learning_rate):
    x_vals = np.linspace(-10, 2, 100)
    y_vals = f(x_vals)
    for i, x_i in enumerate(history):
        plt.clf()
        plt.plot(x_vals, y_vals, label="f(x) = x² + 4x + 4", color='blue')
        plt.scatter(history[:i+1], [f(x) for x in history[:i+1]], color='red')
        plt.scatter(x_i, f(x_i), color='green', s=100, label=f"Iteration {i}")
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.title(f"LR={learning_rate}, Iteration {i}")
        plt.legend()
        plt.pause(0.4)
    plt.show()
```

```python
# Run for different learning rates
learning_rates = [0.1, 0.5, 0.8, 1.0]
starting_point = 0
iterations = 10

for lr in learning_rates:
    print(f"\n=== Learning Rate = {lr} ===")
    minimum, history = gradient_descent(starting_point, lr, iterations)
    print(f"Local minimum at x = {minimum:.4f}, f(x) = {f(minimum):.4f}")
    visualize(history, lr)
```

# **Stochastic Gradient Descent**


```python
# example dataset (each point defines a small loss)
data = [-3, -2.5, -2, -1.5, -1] # made up for elaboration purposes
def dfi(x, ai):
    return 2 * (x - ai)
```

```python
def stochastic_gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    history = [x]
    for i in range(iterations):
        # pick ONE random sample
        ai = np.random.choice(data)
        # gradient from single sample
        grad = dfi(x, ai)

        # SGD update
        x = x - learning_rate * grad

        history.append(x)

        print(
            f"Iteration {i+1}: sample={ai}, "
            f"x={x:.4f}, f(x)={f(x):.4f}"
        )

    return x, history
```

```python

# parameters
starting_point = 0
learning_rate = 0.1
iterations = 10

minimum, history = stochastic_gradient_descent(
    starting_point, learning_rate, iterations
)

print(f"\nApprox minimum at x = {minimum:.4f}, f(x) = {f(minimum):.4f}")
```

```python
def f_plot(x):
    return x**2 + 4*x + 4

def visualize(history, learning_rate):
    x_vals = np.linspace(-10, 2, 200)
    y_vals = f_plot(x_vals)

    plt.figure()

    for i, x_i in enumerate(history):
        plt.clf()

        plt.plot(x_vals, y_vals, label="f(x) = x² + 4x + 4")

        plt.scatter(history[:i+1],
                    [f_plot(x) for x in history[:i+1]])

        plt.scatter(x_i, f_plot(x_i),
                    s=120,
                    label=f"Iteration {i}")

        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.title(f"SGD | LR={learning_rate}, Iteration {i}")
        plt.legend()
        plt.pause(0.4)

    plt.show()
```

```python
starting_point = 0
learning_rate = 0.1
iterations = 15

minimum, history = stochastic_gradient_descent(
    starting_point,
    learning_rate,
    iterations
)

print(f"\nApprox minimum at x = {minimum:.4f}, f(x) = {f(minimum):.4f}")

visualize(history, learning_rate)
```

# **Momentum Gradient Descent function**


In the normal GD update equation of the weights we have noticed that, Learning rate isa always a small value and if the gradient is small too that would lead to phonemena called **Vanishing GD** and to overcome this problem we have came up with new term called **Momentum Term**
It doesn't solve the vanishing GD problem completely but helps the optimizer to go on and continue moving

```python
def f(x):
    return x**2 + 4*x + 4

def df(x):
    return 2*x + 4
```

```python
def momentum_gradient_descent(starting_point,
                              learning_rate,
                              iterations,
                              beta=0.9):

    x = starting_point
    v = 0  # velocity history sum of grad hisotry
    history = [x]

    for i in range(iterations):
        grad = df(x)
        # momentum update
        v = beta * v + grad
        # parameter update
        x = x - learning_rate * v
        history.append(x)
        print(f"Iter {i+1}: x={x:.4f}, f(x)={f(x):.4f}, v={v:.4f}")

    return x, history
```

```python
def visualize(history, learning_rate, optimizer_name="Optimizer"):
    x_vals = np.linspace(-10, 2, 200)
    y_vals = f(x_vals)

    plt.figure()

    for i, x_i in enumerate(history):
        plt.clf()

        # function curve
        plt.plot(x_vals, y_vals, label="f(x)=x²+4x+4")

        # trajectory
        plt.scatter(history[:i+1],
                    [f(x) for x in history[:i+1]])

        # current point
        plt.scatter(x_i, f(x_i),
                    s=120,
                    label=f"Iteration {i}")

        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.title(f"{optimizer_name} | LR={learning_rate}, Iter={i}")
        plt.legend()
        plt.pause(0.4)

    plt.show()
```

```python
starting_point = 0
learning_rate = 0.1
iterations = 15

minimum, history = momentum_gradient_descent(
    starting_point,
    learning_rate,
    iterations,
    beta=0.9
)

print(f"\nMinimum near x = {minimum:.4f}, f(x)={f(minimum):.4f}")

visualize(history, learning_rate, "Momentum GD")
```
